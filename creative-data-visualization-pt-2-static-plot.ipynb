```python
def graph(formula, x_range, formula2, x_range2 # found online for equation drawing https://stackoverflow.com/questions/14000595/graphing-an-equation-with-matplotlib
         ):  # to draw the graph
    y = x_range
    x = formula(y)
    y2 = x_range2
    x2 = formula2(y2)
    plt.plot(x, y)
    plt.plot(x2,y2)
    plt.show()  

def my_formula(x):
    return (-x**5+4*(x**3)/4-3/5)

def my_formula2(x):
    return (x**5-8*(x**3)/4-3/5)

graph(my_formula, np.arange(0,2,0.1), my_formula2, np.arange(-2,0.1,0.1))
```

```python
# to combine the equation and picture
fig, ax = plt.subplots(figsize = (12,10))
ax.imshow(img, extent=[0, 8, 0, 8]) # to correct the axis
y = np.arange(-0.05,2,0.05)
x = my_formula(y)
y2 = np.arange(-2,0.05,0.05)
x2 = my_formula2(y2)
x = [(y+2) for y in x]
y = [(x+2)*1 for x in y]
x2 = [(y+2) for y in x2]
y2 = [(x+2)*1 for x in y2]
ax.plot(x,y, color = '#2B65EC')
ax.plot(x2,y2, color = '#2B65EC')
plt.ylim(0,8)
plt.xlim(0,8)
```
